# Modelo Distribucional de Representación en espacios vectoriales - DSM

```python
import codecs
import json, string, math
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from __future__ import division
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("spanish")          
```

## 1.- Obtención, limpieza y stemming al corpus

* El corpus se encuentra en un archivo de texto plano, cada línea se considera un contexto.
* Para la limpieza y mejor manejo del texto, se realizan las siguientes acciones:
    - Convertir a minúsculas
    - Reemplazar acentos
    - Eliminar signos de puntuación
    - Aplicar proceso de stemming
* Las oraciones se guardan en una lista, las palabras y su conteo en un diccionario


```python
data =[]
Diccionary={}
lineaStemmer=""
TW=0

def remove_punctuation (text):
    text = text.replace("\xe1","a").replace("\xe9","e").replace('\xe8','e').replace("\xed","i").replace("\xf3","o").replace("\xfa","u").replace("\xf1","n")
    return text.translate(None, string.punctuation)

with codecs.open('corpusDSM-PMI.txt','r') as f: # Abrimos el archivo corpusDMS-PMI.txt| corpusPMI.txt
    for line in f: #Recorremos cada línea del archivo
        line= remove_punctuation(line.lower()) #Eliminamos los signos de puntuación, reemplazamos acentos y convertimos a minúsculas
        words= line.split() #Recorremos cada palabra de la línea
        for w in words:
            TW+=1 #Contador de total de palabras
            #w=PorterStemmer().stem_word(w) #Alicamos Stemming
            w = stemmer.stem(w)
            lineaStemmer += w + " "
            if not Diccionary.has_key(w): #Creamos diccionario
                Diccionary[w]=1
            else:
                Diccionary[w]+=1
        lineaStemmer= lineaStemmer[0:len(lineaStemmer)-1]
        data.append(lineaStemmer)#Guardamos la línea con las palabras transformadas
        lineaStemmer=""
```

## 2.- Creación de Diccionario con las palabras ordenadas

```python
Dct = {} #Creamos diccionario que contiene las palabras
Dct = sorted(Diccionary.keys()) #Agregamos y ordenamos las palabras
```

```python
Dct
```

## 3.- Perspectiva de Información Mutua - PMI
Calculamos la probabilidad conjunta **P(x,y)**, tomando como **Contexto** cada oración

```python
PMI= [[0. for j in range(len(Diccionary))]for i in range(len(Diccionary))] #Creamos matriz
#Recorremos el diccionario de palabras
for i in Dct:
    for linea in data: #Recorremos el archivo
        word= linea.split()
        if linea.find(i)>=0:
            for w in word:
                if i!=w:
                    PMI[Dct.index(w)][Dct.index(i)]=+1.0 #Calculamos la probabilidad conjunta P(x,y)
                    #print '%s %s %d ' % (i,w, PMI[Dct.index(w)][Dct.index(i)])
```

## 4.-Calculamos la matriz resultante aplicando la fórmula:
**$PMI(x,y)=\log \frac{P(x,y)}{P(x)P(y)}$**

```python
for i in range(len(Dct)):
    for j in range(len(Dct)):
        if (PMI[i][j]/TW)!=0:
            PMI[i][j]= math.log10((PMI[i][j]/TW)/((Diccionary[Dct[i]]/TW) * (Diccionary[Dct[j]]/TW)))#Calculamos PMI
            #print '[%d %d] %s %s %.4f %d %d'%(i, j, Dct[i], Dct[j],PMI[i][j], Diccionary[Dct[i]], Diccionary[Dct[j]])
```

```python
PMI= np.array(PMI)
```
